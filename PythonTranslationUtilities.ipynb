# Helper Function: Translate to Spanish

```python
def translate_to_spanish(english_input):
    
    from googletrans import Translator
    
    translator = Translator()
    if english_input != "NaN" and english_input != "None":
        output = translator.translate(str(english_input), dest='es')
        return(output.text)
    else:
        return None
```

```python
translate_to_spanish("Hello")
```

```python
translate_to_spanish("NaN")
```

```python
translate_to_spanish("None")
```

# Helper Function: Translate to Any Target Language

```python
def translate_to_target_language(english_input, target_language_keyword):
    
    from googletrans import Translator
    
    translator = Translator()
    if english_input != "NaN" and english_input != "None":
        output = translator.translate(str(english_input), dest=target_language_keyword)
        return(output.text)
    else:
        return None
```

```python
translate_to_target_language("Translation", "es")
```

```python
translate_to_target_language("Translation", "fr")
```
